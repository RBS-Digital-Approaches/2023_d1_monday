People generally thinking of the computational environment as being divided into two major sub-categories of operation and effort:  hardware and software.  Not surprisingly, each of these broad categories is categories actually descirbes a variety of subcategories of efforts, each of which interacts with its sibling categories in ways that directly impact the overall computational work effort. Generally speaking, all computer programmomg languages fall into one of the following categories:

### Assembly Languages

Assembly languages, such as ARM, MIPS, and x86 (of which you've probably never heard) are, the **low-level** computer programmimg languages that work closest to the machine, as we say.  What we mean by this is that they interact directly with the computer hardware itself and are typically encoded directly into chips at the time of manufacture (literally by controlling where electrons are situated on the conductive interior surface of chips) or loaded into a special type of semi-permanent memory in the hardware.  When you install or update a "driver" on your computer part, if not all, of this process is installing or updating assembly code.  Assembly langages are the fastest executing type of computer software because they rund so close to the machine.  A program written in an assembly languaghe will outperform the same programmoimg writtem in another type of programming language by orders of magnitude.  And speed can matter a lot in the current era of big data, as we'll see later in this course.  However, because assembly code is interacting directly with the computer hardware, you need to write a unique assembly program for each unique type of computer hardware, even if the software perfromes the exact same fuction on each system.  In practice, this would mean, for example, that you have to have a uniquely coded version of, for example, Microsoft word, not only for every single make and model of computer on the market but also for every single variation of harware components that exist in each of these systems.  

### Compiled Languages

Compiled languages, such as C, C++, and X-Code, operate one level above Assembly langues.  These languages provide a more generalized programming environment that will work across varrying hardware systems.  Compiled languages process the logic of the programming operation by passing instructions to and recieving feedback from the the Assembly languages that are running closest to the machine.  This communication causes some inneficiency the overal computational process, as a result of which the same process will execute slower if coded in a Compiled language as opposed to an Assembly language.  In most cases, the negative impact of this inneficieny is negligible compared to the effort it would take to write and manage multiple versions of the same program for each required hardware configuration.

Realistically, working in a Compiled langage like C or C++ is the best option for writing programs where speed of execution is paramount.  However, becuase Compiled languages still operate very close to the machine (just one hop away), they still require a lot of configuration and effort to get them running on differnt hardware systems.  This proces, called Compilation, can be quite complex and time consuming, and must be re-executed every time you change your code.  Additionally, because Compiled langages run so close to the machine, each computer program must provide instructions for a lot of basic computational functions, such, for exaple, tellig the computer exactly where on a computer's memory chip to store the information it is holding.  

### Scripting Languages

Scripting languages such as Python, R, and Javascript sit yet one layer further removed from the machine than compiled languages.  Scripting environments provide a real-time, simple environment for coding that requires minimal configuration and no compilation prior to execution.  You just type some code and hit run (or execute a run command).  When you run code writen in a scripting language, the languager interpreter (a compiled app that install on your system) tranlates the code on-the-fly to a set of instructions that is understandable to a library (a collection of code) of compiled code, which takes those instructions and further translates it to a set of low level instructions that is understandble by the Assembly code that runs your computer.  The overwhelming majority of code written by Digital Humanists, specifically, and Data Scientists, in general, is written in Scripting languges, most predominently Python and R because it is so easy to write, debug, and deploy scripting code across systems.  If, however, you work with large datasets or perform computationally complex precsses, the computational load created by translating and passing instructions back and forth through the chain of removal from the actual hardware can slow processes and memory consumption significantly enough to require moving your work further down the processing chain.  If you do a lot of computational work, you will inevitably face this problem at some point.

### 4.  Control Structures: 

Most of the intelligence that computers exhibit is rooted in the application of what are known as Control Structures.  Control Structures are code elements that test the value of designated variables and behave differently based on the value of that variable.  For example, imagine that you are programming a system for automatically distributing food to animals in the Zoo.  Before the system can distribute food to any single animal, it needs to test whether the animal is a cat or a dog and then dispense the correct food accordingly.

Screen Shot 2021-06-15 at 3.21.39 PM.png

***If-Then:*** The most basic Control Structure is the "*If-Then*" statement.  Logically, it is simply a set of instructions to the computer like:

*--If this animal is a Yak Then dispense Yak food*

In Python, we would write this as follows:



```python
# assign the value 'Yak' to the variable 'animal_type'
animal_type = "Yak"

# test whether the value of the 'animal_type' variable
# is equivalent to 'Yak'
if animal_type == "Yak":
   # print a message to screen if the 'animal_type' passes the test
   print("This is a Yak so distribute Yak food.")
```

<font size="1">Note:  Several Python conventions are represented in the above code.  Control Structures (and other multi-line code squenences such as Iterations, which are introduced below, introduced by terminating the first line, which defines the test being performed by the Control Structure, with a colon character (":").  Then, every line of code that should be executed if the test is passed is indented.</font>

We will go over the particular syntax of the Python "If-Then" statement in another lesson.  For now, what is important is to see how the computers form a test of the value of a particular variable in order to make a decision about how to proceed in its execution.

***If-Else:*** One important aspect of Control Structures is that they can present multiple, alternate forks, not simply a binary option as in the example above.  The typical structure for this is the "If-Else" structure. The "If-Else" structure allows us to test multiple conditions and direct execution down multiple paths accordingly.  A suitable type of statement for our automated feeder would do something like the following:

*--If this animal is an Elephant Then dispense Elephant food*

*--Else If this animal is a Gorilla Then dispense Gorilla food*

*--Else If this animal is a Lion Then dispense Lion food*

*--Else If this animal is a Rabbit Then dispense Rabbit food*

*--Else If this animal is a Rhinoceros Then dispense Rhinoceros food*

*--Else If this animal is a Yak Then dispense Yak food*

*--Else skip this animal and send the Zookeeper a notification*




In Python, we would say the same thing like this:

```python
# Define the type of animal to test
animal_type = "Yak"

# Test agains different values and direct execution accordingly
if animal_type == "Elephant":
   print("This is an Elephant so distribute Elephant food.")   
elif animal_type == "Gorilla":
   print("This is a Yak so distribute Gorilla food.")
elif animal_type == "Lion":
   print("This is a Lion so distribute Lion food.")
elif animal_type == "Rabbit":
   print("This is a Rabbit so distribute Rabbit food.")
elif animal_type == "Rhinoceros":
   print("This is a Rhinoceros so distribute Rhinoceros food.")
elif animal_type == "Yak":
   print("This is a Yak so distribute Yak food.")
else:
   print("Unidentified Animal!  Send warning to Zookeeper")
```

<font size="1">Note that Python shortens the conventional "Else-If" syntax to "elif" to facilitate faster coding. Also note how each new control test statement begins at the first cursor position of the line (no indent) and the code to execute if the condition is met is indented under it.

Conditional Statments like the one above create forks in the execution road and the operational complexity of all computer programs is achieved by nesting or otherwise combining different control structures so that the computer program can make decisions of increasing complexity.

### 1. Variable Assignment by Reference or by Value

Most computer programs allow you to assing variables to each other.  For example, you might have a variable labeled `x` that is holding the value "3" as a result of some previous computational process, and you want to assign the new computed value to a new variable labeled `y`.  This assignment can be handled in two ways.  The most intuitive to non proghrammers is called "by Value.". When programs assign a value of one variable to another "by Value" they simply assign whatever value is held in the referenced variable (for example, `x`) to the new value (for exmaple, `y`'). Once that assignment is makde, the to variable have no connection to each other.  You can change the value of `x` or `y` post assignment without having any impacty on the other variable value.

However, when you assign one Python variable to another, Python usually creates a
**reference** to the object. This means that if you assign an object to
multiple variables and change one, all of them will change.

For example, suppose you assign a list to variables `x` and `y`, then change an
element of `y`. Doing so also changes `x`:


```python
x = [1, 2]
y = x
y[0] = 10
x
```

If you want to make a copy of an object, use the object's `copy` method:


```python
x = [1, 2]
y = x.copy()
y[0] = 10
x
```

Reassigning a variable, as opposed to modifying elements, does not affect other
variables. Here's an example:

```python
x = 3
y = x
y = 4
x
```

### 2. Code Re-use

When doing digital work we often need to perform the same activity repeatedly on multiple items or pieces of data.  Iteration is one way to handle this, but if we have to perform this same function multiple times in multiple locations of a workflow, this can create redundancies in which we have to write the same code multiple times.  Imagine, for example, how inefficient and frustrating it would be if you were trying to assemble a piece of furniture and you had to go to the hardware store and get a new wrench every time you needed to tighten a bolt.  

You can think of some chunks of computer code as tools, like a wrench or hammer, that perform a certain type of work, such as, for example, changing all letters in a string to lowercase.  If you were writing code to process a collection of texts you may want to perform a task like this repeatedly at different points in the code, and you wouldn't want to have to re-code this tool every time you had to perform the task.  Consider the workflow shown below, which provides an high-altitude depiction of the steps involved in producing a date-filtered map of the publication places of items in the ESTC and the English Broadside Ballad Archive:

code_reuse_reason.png

As depicted in the above workflow, for each data source we perform the following two general steps:

1.   Acquire the required data;
2.   Clean and normalize the required data;

Finally, once we have cleaned and normalized data from each datasource, we then place the data on a map.

Of particluar note in the workflow are the pink "Normalize Date" and blue "Geocode Publication Place" squares.  Despite the best efforts of librarians and the MARC and TEI guidelines, dates are often represented differently in different records.  As such, the date data that we receive from our sources must be checked and converted so that it is all represented in the same format (a process called data normalization).  The code that we would use to perform date normalization will be the same for both archives because in both cases it must be able to convert from any date format to the desired final date format.

Similarly, since we ultimately want to map the publication location of each item we have collected, for each data source we have to convert the human readable place names that appear in the MARC or TEI to a latitute and longitude that can be located on a map (a process known as geocoding).  As with the date, the same exact process will be used to perform this task regardless of in which archive a place name appears.

Notice that in the linear workflow of our computing process each of the two pink date processing squares and each of the two blue geocoding squares are separated from each other by other processes. This means that the code that performs each of these processes will need to appear twice in our worflow.  This is a waste of programming time (because you have to copy and paste or completely retype the same exact thing twice) and is also very difficult to manage.  If, for example, we later find that there is an error in our date conversion code, we have go and correct the error in two places.

The above is a very simple example, but it illustrates a frequently occuring situation.  There are typically many processess that we need to perform over and over again in multiple places in an overall programatic workflow.  

To address the problem, computer programming languages have developed the idea of *object reuse*, a programming practice that allows you to designate and name sections of code that will be used frequently (like the code used to geocode a place name or normalize a date) and to be able to execute that code frome anywhere else in your program as often as you like.  The workflow below depicts our earlier mapping workflow with an object reuse architecture:

Note that in the above diagram we use the same general workflow, but there is only one each of the pink and blue boxes, which are each referenced from multiple locations in the code rather than having to be duplicated in multiple locations of the code.  

**Categories of Code Re-use:**  Several methods are commonly used to enable code reuse, depending on the scale to which the developer wants to expose code for reuse.  Generally speaking, these fall into two categories:  1) code designed to be reused within a single program or process; and 2) code designed to be imported by other developers into other programs and processes.  You can think of this distinction as referring to code that is designed for local reuse vs. code that is designed for distribution and remote reuse.

**1. Local Code Reuse:**  When we talk about Local Code Resue, we refer to situations where you are writing reusable chunks of code that are unique to your local need. For example, you may notice that certain fields in a set of metadata you are working with need to be reformatted to meet your particualar needs.  Since you will have to do this repeateadly as part of your overall workflow, it makes sense to package this into a "tool" that can applied and re-applied at will, thereby making your local process more efficient and error free.  Such reusable chuncks of code that exist in the context of a large process (often, within the same code file as the rest of the process) or commonly known as Functions, Methods, or Subroutines, depending on the programming language you are working with and some subtle engineering differences.  In a future module you will learn more about these differences and also learn how to write your own functions.  The Code Block below provides a very simple example of the creation and use of a Function in Python.  At this stage in your computational adventure, it is not necessary for you to grasp all of the intricacies of Function writing and calling.  This will be covered in detail in future modules.  We here simply introduce the idea of local code reuse. 

```python
# create a list of animals
animal_type_list = ['Elephant', 'Gorilla', 'Lion', 'Rabbit', 'Rhinoceros', 'Yak']

# define a function that prints "Current animal: ". 
# Just like a Control Structure or Iteration the first line
# defines (def) the function and the subsequent indented lines
# tell the system what to do when the function is called
def my_function():
  print("Current animal:")

# loop through each animal name in our list
for animal_type in animal_type_list :
  # here we tell the coputer to "call" our function,
  # which is computerspeak for 'run the executable
  # that is defined in the function
  my_function()
  # And here we print the name of the animal
  print(animal_type)
```

<font size="1">Note that the function needs to be defined before it can be called, which which is why it appears at the top of our code, before we need it.</font>

The above example is a fairly simplistic one that really offers very little by way of operational efficiency.  However, it illustrates the basic mechanism of code reuse through the creation of functions.  In later modules you will learn more about how to create and use your own functions.

**2. Distributed Code Reuse:** As digital bibliographers we spend a great deal of time writing code and functions to solve very unique local problems related to our specific research questions.  But sometimes we, or another developer, may develop a process that can be used by others in novel ways to solve their unique coding and research problems.  For example, later in this course we will learn about a proces called Topic Modeling, which uses an algorithm called Latent Dirichlet Allocaion (LDA) to computationally identify the semantic topics that appear in a collection of texts.  LDA is a very complicated mathematical algorithm and it takes several thousands of lines of code to program the LDA calculus into a set of programmatic instructions for the computer.

If every person who wanted to use LDA for Topic Modeling texts were required to write the code necessary to convert LDA's calculus to code, it would significanly impact the community's ability to use the algorithm for specific research purposes.  Happily, this isn't necessary, because the the computer scientists who developed the LDA method have create a complete LDA 'package' and made it available to other developers to 'import' into their own code and worflows.  Because they have packaged their tool for distributed reuse, any developer can use LDA with only three lines of code, as opposed to thousands. # VI. Getting Help

Learning and using a language is hard, so it's important to know how to get
help. The first place to look for help is Python's built-in documentation. In
the console, you can access the help pages with the `help` function.

There are help pages for all of Python's built-in functions, usually with the
same name as the function itself. So the code to open the help page for the
`round` function is:

```python
help(round)
```

For functions, help pages usually include a brief description and a list of
parameters and default arguments. For instance, the help page for `round` shows
that there are two parameters `number` and `ndigits`. It also says that
`ndigits=None`, meaning the default argument for `ndigits` is the special
`None` value, which you'll learn more about on day 2.

There are also help pages for other topics, such as built-in operators and
modules (you'll learn more about modules in {numref}`modules-packages`). To
look up the help page for an operator, put the operator's name in single or
double quotes. 

Python treats anything inside single or double quotes as literal text rather
than as an expression to evaluate. In programming jargon, a piece of literal
text is called a **string**. You can use whichever kind of quotes you prefer,
but the quote at the beginning of the string must match the quote at the end. 

For example, this code opens the help page for the arithmetic operators:

```python
help("+")
```

It's always okay to put quotes around the name of the page when you use `help`,
but they're only required if the name contains non-alphabetic characters. So
`help(abs)`, `help('abs')`, and `help("abs")` all open the documentation for
`abs`, the absolute value function.

You can also browse the Python documentation [online][pydocs]. This is a good
way to explore the many different functions and data structures built into
Python. If you do use the online documentation, make sure to use the
documentation for the same version of Python as the one you have. Python
displays the version each time you open a new console, and the online
documentation shows the version in the upper left corner.

[pydocs]: https://docs.python.org/3/

Sometimes you might not know the name of the help page you want to look up. In
that case it's best to use an online search engine. When you search for help
with Python online, include "Python" as a search term.

# VII.  When Things Go Wrong

As a programmer, sooner or later you'll run some code and get an error message
or result you didn't expect. Don't panic! Even experienced programmers make
mistakes regularly, so learning how to diagnose and fix problems is vital.

Try going through these steps:

1. If Python printed a warning or error message, read it! If you're not sure
   what the message means, try searching for it online.
2. Check your code for typographical errors, including incorrect
   capitalization, whitespace, and missing or extra commas, quotes, and
   parentheses.
3. Test your code one line at a time, starting from the beginning. After each
   line that assigns a variable, check that the value of the variable is what
   you expect. Try to determine the exact line where the problem originates
   (which may differ from the line that emits an error!).

If none of these steps help, try asking online. [Stack Overflow][stacko] is a
popular question and answer website for programmers. Before posting, make sure
to read about [how to ask a good question][goodq].

[stacko]: https://stackoverflow.com/
